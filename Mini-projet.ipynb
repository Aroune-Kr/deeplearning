```python
np.random.seed(0)
df = {
    'Pays': np.random.choice(['France', 'Allemagne', 'Espagne', 'Italie', 'Norvège'], 100),
    'Année': np.random.choice([2018, 2019, 2020, 2021], 100),
    'Consommation (%)': np.random.uniform(10, 60, 100)
}
df = pd.DataFrame(df)
```

```python
df
```

```python
# Filtrer les données pour un pays spécifique, par exemple la France
df_france = df[df['Pays'] == 'France'].sort_values('Année')

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_france, x='Année', y='Consommation (%)', marker='o')
plt.title('Évolution de la Consommation d’Énergie Renouvelable en France')
plt.xlabel('Année')
plt.ylabel('Consommation d’Énergie Renouvelable (%)')
plt.show()
```

```python
# Sélectionner les données pour l'année 2020
df_2020 = df[df['Année'] == 2020]

plt.figure(figsize=(10, 6))
sns.histplot(df_2020['Consommation (%)'], bins=20, kde=True)
plt.title('Distribution de la Consommation d’Énergie Renouvelable en 2020')
plt.xlabel('Consommation d’Énergie Renouvelable (%)')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Sélectionner quelques pays pour la comparaison
pays_choisis = ['France', 'Allemagne', 'Espagne']
df_choisis = df[df['Pays'].isin(pays_choisis)]

plt.figure(figsize=(12, 8))
sns.lineplot(data=df_choisis, x='Année', y='Consommation (%)', hue='Pays', marker='o')
plt.title('Comparaison de la Consommation d’Énergie Renouvelable entre Pays')
plt.xlabel('Année')
plt.ylabel('Consommation d’Énergie Renouvelable (%)')
plt.legend(title='Pays')
plt.show()
```

```python
# Créer un pivot table pour la heatmap
heatmap_data = df.pivot_table(values='Consommation (%)', index='Pays', columns='Année', aggfunc=np.mean)

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Heatmap de la Consommation d’Énergie Renouvelable par Pays et Année')
plt.xlabel('Année')
plt.ylabel('Pays')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Année', y='Consommation (%)')
plt.title('Distribution de la Consommation d’Énergie Renouvelable par Année')
plt.xlabel('Année')
plt.ylabel('Consommation d’Énergie Renouvelable (%)')
plt.show()
```

```python
# Supposons que vos données pour l'année spécifique sont stockées dans df_2021
# Nous allons d'abord trier ces pays par leur consommation d'énergie renouvelable
df_sorted = df[df['Pays'] == 2000].sort_values('Consommation (%)', ascending=False)

# Ensuite, sélectionner les 10 premiers pays
top_10_pays = df_sorted.head(10)

# Création du graphique
plt.figure(figsize=(12, 8))
sns.barplot(x='Indicator Name', y='Pays', data=top_10_pays, palette='coolwarm')
plt.title('Top 10 des pays par consommation d’énergie renouvelable en 2021')
plt.xlabel('Consommation d’énergie renouvelable (%)')
plt.ylabel('')
plt.show()
```

```python
df = df.nlargest(10, '2000')

# Créer le graphique à barres
plt.figure(figsize=(10, 6))
sns.barplot(x='2020', y='Country Name', data=df, palette='viridis')
plt.title('Top 10 des Pays par Consommation d’Énergie Renouvelable en 2020')
plt.xlabel('Consommation d’Énergie Renouvelable (%)')
plt.ylabel('Pays')
plt.show()
```
